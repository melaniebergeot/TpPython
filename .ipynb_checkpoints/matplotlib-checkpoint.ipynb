# travailler avec differentes figueres et axes 

Toutes les commandes de tra√ßage s'appliquent aux axes actuels. La fonction gca () renvoie les axes actuels (une instance matplotlib.axes.Axes), et gcf () renvoie la figure courante (instance matplotlib.figure.Figure). Normalement, vous n'avez pas √† vous en soucier, car tout est pris en charge dans les coulisses. Voici un script pour cr√©er deux subplots.

```python
import numpy as np
import matplotlib.pyplot as plt

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

figure() est une commande optionnelle car figure(1) peut √™tre cr√©er par d√©faut. Comme subplot (111)npeut √™tre cr√©er par d√©faut si on ne sp√©cifie pas manuellement les axes par defaut. 

La commande subplot() specifies numrows, numcols, fignum where fignum ranges from 1 to numrows*numcols. 


The commas in the subplot command are optional if numrows*numcols<10. So subplot(211) is identical to subplot(2, 1, 1). You can create an arbitrary number of subplots and axes. If you want to place an axes manually, i.e., not on a rectangular grid, use the axes() command, which allows you to specify the location as axes([left, bottom, width, height]) where all values are in fractional (0 to 1) coordinates. See pylab_examples example code: axes_demo.py for an example of placing axes manually and pylab_examples example code: subplots_demo.py for an example with lots of subplots.

You can create multiple figures by using multiple figure() calls with an increasing figure number. Of course, each figure can contain as many axes and subplots as your heart desires:





```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

A SAVOIR: la m√©moire requise pour une figure n'est pas compl√®tement lib√©r√©e jusqu'√† ce que la figure soit explicitement ferm√©e avec close ().

Supprimer toutes les r√©f√©rences √† la figure, et / ou utiliser le gestionnaire de fen√™tres pour quitter la fen√™tre dans laquelle la figure appara√Æt √† l'√©cran, ne suffit pas, car pyplot maintient les r√©f√©rences internes jusqu'√† ce que close () soit appel√©.

## Travailler avec du texte: 

La commande texte() peut √™tre utilis√© pour ajouter du texte a un endroit pr√©cis. 
xlabel(),ylabel() et title() sont utilis√© pour ajouter du texte
voir l'exemple si dessous : 

```python
import numpy as np
import matplotlib.pyplot as plt

# Fixing random state for reproducibility
np.random.seed(19680801)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

on peut modifier les propri√©t√©s  passant des arguments de mot-cl√© dans les fonctions de texte ou en utilisant setp ()

```python
t = plt.xlabel('my data', fontsize=14, color='red')
```

# Annoter du texte

Les utilisations de la commande basic text () ci-dessus placent le texte √† une position arbitraire sur les axes.

Une utilisation courante du texte est d'annoter certaines caract√©ristiques du trac√©, et la m√©thode annotate () fournit une fonctionnalit√© d'aide pour faciliter les annotations.
Dans une annotation, il y a deux points √† consid√©rer: l'emplacement annot√© repr√©sent√© par l'argument xy et l'emplacement du texte xytext. Ces deux arguments sont des tuples (x, y).


```python
import numpy as np
import matplotlib.pyplot as plt

ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)

plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            )

plt.ylim(-2,2)
plt.show()
```

Dans cet exemple, xy et xytext sont en coordonnes de donn√©es. 
D'autres exemples existe comme pylab_examples: annotation_demo.py. : https://matplotlib.org/examples/pylab_examples/annotation_demo.html#pylab-examples-annotation-demo


## Logarithmique et autres axes non lin√©aires 

matplotlib.pyplot prend en charge non seulement les √©chelles d'axes lin√©aires, mais aussi les √©chelles logarithmiques et "logit". 
Couramment utilis√© si les donn√©es couvrent de nombreux ordres de grandeur.
On change l'√©chelle comme ceci : 

plt.xscale ('log')

Un exemple de quatre graphiques avec les m√™mes donn√©es et diff√©rentes √©chelles pour l'axe y est pr√©sent√© ci-dessous: 


```python
import numpy as np
import matplotlib.pyplot as plt

from matplotlib.ticker import NullFormatter  # useful for `logit` scale

# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)


# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)


# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Format the minor tick labels of the y-axis into empty strings with
# `NullFormatter`, to avoid cumbering the axis with too many labels.
plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)
```
