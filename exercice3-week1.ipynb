1. téléchargement et nettoyage du fichier (suppression du HTML)

J'ai décidé d'utilisé BeautifulSoup afin de telecharger le fichier

```python
from bs4 import BeautifulSoup #importation des blibliothèques
import urllib
from urllib.request import urlopen
```

```python
url = 'https://archive.org/stream/lesmiserables01hugogoog/lesmiserables01hugogoog_djvu.txt'## on stocke l'url
page = urlopen(url) ##on recupere le contenu de la page web 
bsobj = BeautifulSoup(page.read(), 'html.parser') ##va faire la distinction entre les balises 
skudiv = bsobj.find("pre").get_text() ##recupère toutes les donnes presente dans la balise pre
texte = skudiv.split('Cd?.') ##saute la partie google jusqu'a cd? 
texte_sel = texte[1].split()##partie qui nous interesse
print(texte_sel)
```

```python
texte_sel_2 = [] #creation d'une array vide
for valeur in texte_sel: 
    for ch in [',', '.', ';', '!', ',','-','&','#', '_', '*','?','.','^','¨']: ## selectionne les valeurs avec des charactères genant 
        if ch in valeur:
            valeur_tmp = valeur.lower().replace(ch, '') #met en minuscule et remplace les char gènant par des espaces
    if len(valeur_tmp) != 0: ## supprime les valeurs ou il n'y a rien 
        texte_sel_2.append(valeur.lower().replace(ch, '')) #stocke dans texte_sel_2
```

2. calcul des occurences de chaque mot

```python
compte = {}.fromkeys(set(texte_sel_2),0)  #creation d'un compteur
for valeur in texte_sel_2: 
    compte[valeur] += 1
print (compte)
```

3. affichage et analyse de la distribution

```python
tri = sorted(compte.items(), key=lambda x:x[1]) ##trier tout les items du dictionnaire du plus petit au plus grand
for i in tri :
    print (i)
```
